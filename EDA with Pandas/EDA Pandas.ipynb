## LETS ANSWER SOME QUESTIONS ABOUT THIS DATA

```python
#What are the locations with the fastest roller coasters (minimum of 10) ?
```

```python
#menim ilk aglima gelen yol budur amma Burda problem odur ki orada 5 ci indexde Other yazilib, ve bu problemdi onu silmek lazimdir
df[['Location','Speed_mph']].sort_values(by='Speed_mph',ascending=False).reset_index(drop=True).head(10)
```

```python
#Artiq other yoxdur
df[df['Location'] != 'Other'][['Location','Speed_mph']].sort_values(by='Speed_mph',ascending=False).reset_index(drop=True).head(10)
```

```python
#Ikinci ve daha rahat yolu  - WOW bu cox daha yaxsi ve rahatdir, Biraz SQL Vibe verir
df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >=10') \ #burdaki 2 ci query sql deki having kimi olub
    .sort_values(by = 'mean')
```

```python
#Ikinci ve daha rahat yolu  - WOW bu cox daha yaxsi ve rahatdir, Biraz SQL Vibe verir
df.query('Location != "Other"') \
  .groupby('Location')['Speed_mph'] \
  .agg(['mean', 'count']) \
  .query('count >= 10') \
  .sort_values(by='mean')['mean'] \
  .plot(kind='barh', title='Average Coast Speed by Location')
  #bu ['mean'] e qeder olan hisse bize series verir indexdede Location olur
```

```python
df.query('Location != "Other"') \
  .groupby('Location')['Speed_mph'] \
  .agg(['mean', 'count']) \
  .query('count >= 10') \
  .sort_values(by='mean')['mean']
```
